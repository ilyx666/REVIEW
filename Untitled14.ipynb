мы видим, что атрибуты, отражающие полярные координаты имеют тип object. Для большей мобильности в исследовании переведем их в float

```python
df['start_lat'] = df['start_lat'].str.replace(',', '.')
df['start_lon'] = df['start_lon'].str.replace(',', '.')
df['driver_start_lat'] = df['driver_start_lat'].str.replace(',', '.')
df['driver_start_lon'] = df['driver_start_lon'].str.replace(',', '.')
df['end_lat'] = df['end_lat'].str.replace(',', '.')
df['end_lon'] = df['end_lon'].str.replace(',', '.')

df['start_lat'] = pd.to_numeric(df['start_lat'], errors='coerce')
df['start_lon'] = pd.to_numeric(df['start_lon'], errors='coerce')
df['driver_start_lat'] = pd.to_numeric(df['driver_start_lat'], errors='coerce')
df['driver_start_lon'] = pd.to_numeric(df['driver_start_lon'], errors='coerce')
df['end_lat'] = pd.to_numeric(df['end_lat'], errors='coerce')
df['end_lon'] = pd.to_numeric(df['end_lon'], errors='coerce')
```

Экстрактируем дату и время из начала и окончания поездки, чтобы дальше нам было проще управлять этими параметрами

```python
df['start_at'] = pd.to_datetime(df['start_at'])
df['end_at'] = pd.to_datetime(df['end_at'])

df['start_date_at'] = df['start_at'].dt.date
df['start_time_at'] = df['start_at'].dt.time
df['end_date_at'] = df['end_at'].dt.date
df['end_time_at'] = df['end_at'].dt.time
```

# Оценки распределения
Исходя из построенных графиков распределения атрибутов нельзя однозначно сделать вывод о типе распределения объектов. За исключением координат начала и окончания поездки. Там прослеживается четкое Гауссово распределение. Также эта закономерность прослеживается и в описательной статистике, приведенной в пункте 1.

```python
import statsmodels.api as sm

# Сортировка данных для построения ECDF
start_lat_sorted = np.sort(df['start_lat'])
start_lon_sorted = np.sort(df['start_lon'])

# Создание ECDF
ecdf_lat = sm.distributions.ECDF(start_lat_sorted)
ecdf_lon = sm.distributions.ECDF(start_lon_sorted)

plt.figure(figsize=(10, 6))
plt.step(ecdf_lat.x, ecdf_lat.y, label='Широта (start_lat)')
plt.xlabel('Координата')
plt.ylabel('Вероятность')
plt.title('Эмпирическая функция распределения (ECDF) для координат')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.step(ecdf_lon.x, ecdf_lon.y, label='Долгота (start_lon)')
plt.xlabel('Координата')
plt.ylabel('Вероятность')
plt.title('Эмпирическая функция распределения (ECDF) для координат')
plt.legend()
plt.grid(True)
plt.show()
```

```python
end_lat_sorted = np.sort(df['end_lat'])
end_lon_sorted = np.sort(df['end_lon'])

# Создание ECDF
ecdf_lat = sm.distributions.ECDF(end_lat_sorted)
ecdf_lon = sm.distributions.ECDF(end_lon_sorted)

plt.figure(figsize=(10, 6))
plt.step(ecdf_lat.x, ecdf_lat.y, label='Широта (end_lat)')
plt.xlabel('Координата')
plt.ylabel('Вероятность')
plt.title('Эмпирическая функция распределения (ECDF) для координат')
plt.legend()
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.step(ecdf_lon.x, ecdf_lon.y, label='Долгота (end_lon)')
plt.xlabel('Координата')
plt.ylabel('Вероятность')
plt.title('Эмпирическая функция распределения (ECDF) для координат')
plt.legend()
plt.grid(True)
plt.show()
```

Собственно, исходя из эмпиричесих функций распределения мы и подтверждаем выдвинутую гипотезу о типе распределения

# Кластеризация

Руководствуясь логикой и бизнес-задачей я бы кластеризовал данные 2 раза.
1. Для определения категорий пользователей - пользователей, по частоте совершения поездки (для внедрения возможной бонусной системы или различных скидок).
2. Для определения районов города. Это сделал бы для установки или настройки различных коэффициентов и комиссий компании для более точного логистического управления водителями


```python
# @title Определение оптимального количества кластеров
from sklearn.cluster import KMeans
num_clusters = range(1, 10)
kmeans = [KMeans(n_clusters=i) for i in num_clusters]


score = [kmeans[i].fit(df[['start_lat', 'start_lon']]).inertia_ for i in range(len(kmeans))]

# Построение графика метода локтя
plt.plot(num_clusters, score)
plt.xlabel('Количество кластеров')
plt.ylabel('Сумма квадратов расстояний')
plt.title('Метод локтя')
plt.show()
```

По данной эвристике оптимальное количество кластеров равно 4. Используем это количество для кластеризации наших данных

```python
n_clusters = 4

kmeans = KMeans(n_clusters=n_clusters, random_state=42)

kmeans.fit(df[['start_lat', 'start_lon']])

df['kmeans_cluster'] = kmeans.labels_

plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='start_lon', y='start_lat', hue='kmeans_cluster', palette='viridis')
plt.title('K-means Clustering')
plt.xlabel('Долгота (start_lon)')
plt.ylabel('Широта (start_lat)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

В данном случае кластеризация райнов города происходит по координатам старта поездки, что обуславливает географическое разделение районов.
Исходя из координат ширины и долготы, а также меток кластеров можно сделать вывод о распределении районов:
0. Район Jesus Maria - набережна и культурный центр города
1. Район Santiago de Surco - спальный район
2. Los Libertadores - спальный район, с вкраплением бизнес районов
3. Район археологической зоны Pirca Pirca и Laguna Piticocha, которые являются популярными туристическими маршрутами

```python
from sklearn.cluster import DBSCAN

# Создание экземпляра модели DBSCAN
dbscan = DBSCAN(eps=0.1, min_samples=5)

# Обучение модели на данных
dbscan.fit(df[['start_lat', 'start_lon']])

# Предсказание кластеров
df['dbscan_cluster'] = dbscan.labels_

# Визуализация результатов
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='start_lon', y='start_lat', hue='dbscan_cluster', palette='viridis')
plt.title('DBSCAN Clustering')
plt.xlabel('Долгота (start_lon)')
plt.ylabel('Широта (start_lat)')
plt.legend(title='Cluster')
plt.grid(True)
plt.show()
```

DBScan классифицировал точки как выбросы, если они не принадлежат ни к одному кластеру. В таком случае, необходимо масштабировать признаки, но В контексте данной задачи в этом нет необходимости. Данный алгоритм кластеризации нам не подходит из-за очевидной несбалансированности кластеров

```python
from sklearn.cluster import AgglomerativeClustering

agglomerative_cluster = AgglomerativeClustering(n_clusters=4)

clusters = agglomerative_cluster.fit_predict(df[['start_lat', 'start_lon']])

df['cluster'] = clusters

print(df['cluster'].value_counts())

print(df.head())
```

Иерархическая кластеризация так же показывает явный дисбаланс кластеров.

**В качестве алгоритма кластеризации мы выберем модель Kmeans, так как самые адекватные результаты были получены именно с использованием данного алгоритма. 3 кластер - археологическая зона и зона лагуны явно выделяются меньшим количеством объектов, но в данном случае я считаю такое допущение нормальным, так как географически этот кластер наиболее удаленный от города Лима, в котором и были собраны данные**
