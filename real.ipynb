```python
from sklearn.feature_extraction.text import CountVectorizer
vectorizer = CountVectorizer()
X = vectorizer.fit_transform(df.Rewiew)
X
```

```python
new_X = pd.concat([pd.DataFrame(X.toarray()), df[['Rating', 'amount Photo', 'LenReview']]], axis=1)
```

```python
y = df.isReal
```

```python
new_X.columns = new_X.columns.astype(str)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(new_X , y, test_size=0.2, random_state=42)
```

```python
new_X
```

```python
from sklearn.linear_model import LogisticRegression
clf = LogisticRegression(max_iter = 1000, random_state=0, C=10).fit(X_train, y_train)
```

```python
x_predict = clf.predict(X_test)
```

```python
import numpy as np

from sklearn.metrics import accuracy_score
print(accuracy_score(y_test, x_predict))
```

```python
lr = LogisticRegression()
from sklearn.model_selection import GridSearchCV
param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100, 1000]}

# создание экземпляра класса GridSearchCV
grid_search = GridSearchCV(lr, param_grid, cv=5)

# запуск поиска наилучших гиперпараметров
grid_search.fit(X_train, y_train)

# вывод наилучших гиперпараметров
print("Best hyperparameters: ", grid_search.best_params_)

# вывод значения метрики на тестовой выборке с наилучшими гиперпараметрами
print("Accuracy on test set with best hyperparameters: ", grid_search.score(X_test, y_test))
```

```python
from sklearn.ensemble import GradientBoostingClassifier
gb_clf = GradientBoostingClassifier(n_estimators=8957, learning_rate=0.1, max_depth=3, random_state=42)

# Обучение модели на обучающей выборке
gb_clf.fit(X_train, y_train)

# Получение предсказаний на тестовой выборке
y_pred = gb_clf.predict(X_test)

# Вычисление точности модели
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```
