```python
import re
import string
import nltk
from nltk.corpus import stopwords
from pymystem3 import Mystem

# –∑–∞–≥—Ä—É–∂–∞–µ–º —Å—Ç–æ–ø-—Å–ª–æ–≤–∞ –¥–ª—è —Ä—É—Å—Å–∫–æ–≥–æ —è–∑—ã–∫–∞
nltk.download('stopwords')
stop_words = set(stopwords.words('russian'))

# —Å–æ–∑–¥–∞–µ–º –æ–±—ä–µ–∫—Ç –ª–µ–º–º–∞—Ç–∏–∑–∞—Ç–æ—Ä–∞
m = Mystem()

# —Ñ—É–Ω–∫—Ü–∏—è –¥–ª—è –ø—Ä–µ–¥–æ–±—Ä–∞–±–æ—Ç–∫–∏ —Ç–µ–∫—Å—Ç–∞
def preprocess_text(text):
    # —É–¥–∞–ª—è–µ–º —Å—Å—ã–ª–∫–∏ –∏ —Ö—ç—à—Ç–µ–≥–∏
    text = re.sub(r'http\S+', '', text)
    text = re.sub(r'#\S+', '', text)
    # —É–¥–∞–ª—è–µ–º –∑–Ω–∞–∫–∏ –ø—Ä–µ–ø–∏–Ω–∞–Ω–∏—è
    text = text.translate(str.maketrans('', '', string.punctuation))
    # –ø—Ä–∏–≤–æ–¥–∏–º –∫ –Ω–∏–∂–Ω–µ–º—É —Ä–µ–≥–∏—Å—Ç—Ä—É
    text = text.lower()
    # –ª–µ–º–º–∞—Ç–∏–∑–∏—Ä—É–µ–º —Ç–µ–∫—Å—Ç
    lemmas = m.lemmatize(text)
    # —É–¥–∞–ª—è–µ–º —Å—Ç–æ–ø-—Å–ª–æ–≤–∞
    filtered_lemmas = [lemma for lemma in lemmas if lemma.strip() and lemma not in stop_words]
    # –æ–±—ä–µ–¥–∏–Ω—è–µ–º –ª–µ–º–º—ã –≤ —Å—Ç—Ä–æ–∫—É
    filtered_text = ' '.join(filtered_lemmas)
    return filtered_text

# –ø—Ä–∏–º–µ–Ω—è–µ–º —Ñ—É–Ω–∫—Ü–∏—é –∫ –∫–∞–∂–¥–æ–º—É —ç–ª–µ–º–µ–Ω—Ç—É —Å—Ç–æ–ª–±—Ü–∞ Rewiew
for idx in range(len(df['Rewiew'])):
    df['Rewiew'][idx] = preprocess_text(df['Rewiew'][idx])
```
